```
dataset_name = 'healthtapQAs'

with open('gdrive/My Drive/Colab Notebooks/medical-question-answer-data/%s.json'%dataset_name) as file:
  ds = json.load(file)
  file.close()
print('dataset length: ', len(ds))
```

```
from bert_serving.client import BertClient
bc = BertClient()
```

```
output = open('gdrive/My Drive/Colab Notebooks/mqa-biobert/%s.csv'%dataset_name, "w")
writer = csv.writer(output)
writer.writerow(['question', 'answer', 'question_bert', 'answer_bert'])
for i in tqdm(range(len(ds)), mininterval=30, maxinterval=60):
  sample = ds[i]
  question, answer = sample['question'], sample['answer']
  question_bert = bc.encode([question])
  answer_bert = bc.encode([answer])
  writer.writerow([question, answer, question_bert, answer_bert])
output.close()
```
